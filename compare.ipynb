## 将官方list中的ip地址中的多余的0去掉，转换成标准ip地址

```python
true_srcip = []
for i in truelist['src_ip']:
    trasformedsrc = ".".join([str(int(s)) for s in i.split(".")])
    true_srcip.append(trasformedsrc)
true_dstip = []
for i in truelist['dst_ip']:
    transformeddst = '.'.join([str(int(s)) for s in i.split(".")])
    true_dstip.append(transformeddst)
```

```python
true_src_ip = pd.DataFrame(true_srcip)
true_src_ip.columns = ['src_ip']
true_dst_ip = pd.DataFrame(true_dstip)
true_dst_ip.columns = ['dst_ip']
```

```python
print(true_src_ip.shape,true_dst_ip.shape)
```

## 将src_ip,dst_ip插入到truelist正确位置上

```python
truelist_1 = truelist.drop(['src_ip', 'dst_ip'], axis=1, inplace=False)
```

```python
truelist_1 = pd.concat([truelist_1, true_src_ip, true_dst_ip], axis=1)
```

```python
truelist_1.shape
```

```python
df_src = truelist_1.src_ip
truelist_1 = truelist_1.drop('src_ip', axis=1)
truelist_1.insert(2, 'src_ip', df_src)
```

```python
df_dst = truelist_1.dst_ip
truelist_1 = truelist_1.drop('dst_ip', axis=1)
truelist_1.insert(3, 'dst_ip', df_dst)
```
